```python
import pandas as pd
```

```python

df = pd.read_csv('../outputs/capacityvpresence.csv')
df.drop(columns=df.columns[0], axis=1, inplace=True)
df.head()
```

```python
# Adding macro-area information
def categorize_region(region):
    region_lower = region.lower()
    if region_lower in ['trentino alto adige', 'veneto', 'friuli venezia giulia', 'emilia romagna']:
        return 'nord-est'
    elif region_lower in ["valle d'aosta", 'liguria', 'lombardia', 'piemonte']:
        return 'centro'
    elif region_lower in ['toscana', 'umbria', 'marche', 'lazio']:
        return 'centro'
    elif region_lower in ['abruzzo', 'molise', 'campania', 'puglia', 'basilicata', 'calabria']:
        return 'sud'
    elif region_lower in ['sicilia', 'sardegna']:
        return 'isole'

df['Region Group'] = df['Region'].apply(categorize_region)
df.sample(10)
```

```python
df['Occupancy Rate'] = (df['Present'] / df['Capacity'])- 1

df.head()
```

```python
df.to_csv('../outputs/occupancy_rates.csv')
```

```python
df_rate_pivot = df.pivot(index='Date', columns='Region', values='Occupancy Rate')
df_rate_pivot
```

```python
df_rate_pivot.to_csv('../outputs/occupancy_rates_pivot.csv')
```
